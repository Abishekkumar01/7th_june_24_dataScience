# # scatter plot 


```python
price = list(df['price'].values)
age = list(df['age'].values)


# note: 2 columns and columns must be numerical
plt.scatter(age,price,color='r',marker='o') 
plt.xlabel("AGE")
plt.ylabel("PRICE")
plt.title("AGE Vs PRICE")
plt.show()
```

#  barplot 


```python
brand = list(df['brand'].value_counts().keys())   # categorical 
no_of_count = list(df['brand'].value_counts().values)  # numerical 

plt.figure(figsize=(14,7))   # width , height 
plt.bar(brand,no_of_count , color='green',ec='g',lw=4)  # ec = edge color  ,lw=line width , ls= dashed , dotted 
plt.xlabel('BRANDS')
plt.ylabel('nO_OF COUNTS')
plt.title("No. of BIKE in BRand")

# plt.xticks(rotate=60)
plt.xticks(rotation=90)
plt.show()
# barplot 
```

```python
bullet['owner'].value_counts()
# pie plot  1 numerical categorical 
```

```python
ownership = list(bullet['owner'].value_counts().keys()) 
no_of_count = list(bullet['owner'].value_counts().values)
```

```python
ownership
```

```python
no_of_count
```

#  pie plot 


```python
plt.figure(figsize=(8,8))
plt.pie(no_of_count,labels=ownership,autopct='%1.2f%%')
plt.show()
```

```python
ownership = list(df['owner'].value_counts().keys()) 
no_of_count = list(df['owner'].value_counts().values)
```

```python
ownership = list(df['owner'].value_counts().keys()) 
no_of_count = list(df['owner'].value_counts().values)
plt.figure(figsize=(8,8))
plt.pie(no_of_count,labels=ownership,autopct='%1.2f%%')
plt.show()
```

```python
country = ['india','china','japan','pakistan']
population = [900,700,500,350]
explosion = [0,0.2,0,0.4]
plt.figure(figsize=(8,8))
plt.pie(population,labels=country,autopct='%1.2f%%',wedgeprops={'ec':'k'},explode=explosion,shadow=True)
plt.savefig('my_pie_plot')
plt.show()
```
